# Test Varients 

Two tests should be conducted. 

1. _*Guardians Test*_: a (parents & teachers) test where the parent *and* teacher take the assessment on behalf of the student. 
2. _*Scholars Test*_: a test that should be taken by the candidate. It should include personalisation & freedom that does not neccessarily indicate personality, but rather allows the student to feel engaged.
3. _*Peers Test*_: Possibly a third test, taken by the students peers. (to come in a later session)


The different perspectives are combined form the students personality diagnostics.
